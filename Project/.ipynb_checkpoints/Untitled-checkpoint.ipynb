```python
import pandas as pd
import matplotlib.pyplot as plt

# Load player data
player_data = pd.read_csv('player_data.csv')

# Plot histogram of player heights
plt.figure(figsize=(10, 6))
plt.hist(player_data['height'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Player Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load match data
match_data = pd.read_csv('match_data.csv')

# Extract home and away goals
home_goals = match_data['home_team_goal']
away_goals = match_data['away_team_goal']

# Create an overlay histogram
plt.figure(figsize=(10, 6))
plt.hist(home_goals, bins=range(0, max(home_goals.max(), away_goals.max())), alpha=0.5, label='Home Team Goals', color='green')
plt.hist(away_goals, bins=range(0, max(home_goals.max(), away_goals.max())), alpha=0.5, label='Away Team Goals', color='blue')

# Add titles and labels
plt.title('Distribution of Goals by Home and Away Teams')
plt.xlabel('Number of Goals')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Load team attributes data
team_attr = pd.read_csv('team_attr.csv')

# Convert date column to datetime
team_attr['date'] = pd.to_datetime(team_attr['date'])

# Plot build-up play speed over time
plt.figure(figsize=(12, 6))
for team_id in team_attr['team_api_id'].unique():
    team_data = team_attr[team_attr['team_api_id'] == team_id]
    plt.plot(team_data['date'], team_data['buildUpPlaySpeed'], marker='o', linestyle='-', alpha=0.5)

plt.title('Build Up Play Speed Over Time')
plt.xlabel('Date')
plt.ylabel('Build Up Play Speed')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load player stats data
player_stats = pd.read_csv('player_stats.csv')

# Select only numerical columns
numerical_columns = player_stats.select_dtypes(include=['float64', 'int64']).columns

# Set up the figure for subplots
num_cols = len(numerical_columns)
fig, axes = plt.subplots(nrows=(num_cols // 3) + 1, ncols=3, figsize=(20, num_cols * 2))
axes = axes.flatten()

# Create a histogram for each numerical column
for i, column in enumerate(numerical_columns):
    axes[i].hist(player_stats[column].dropna(), bins=20, color='skyblue', alpha=0.7)
    axes[i].set_title(f'Distribution of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

# Remove empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
plt.savefig("Hello.jpeg")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
player_stats = pd.read_csv('player_stats.csv')
team_data = pd.read_csv('team_data.csv')

# Merge the datasets (this is an assumption and may not be accurate)
merged_data = player_stats.merge(team_data, left_on='id', right_on='team_api_id', how='left')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(merged_data['overall_rating'].dropna(), bins=20, kde=True)
plt.title('Distribution of Player Overall Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Count')
plt.show()
```

```python
top_teams = merged_data.groupby('team_long_name')['overall_rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_teams.plot(kind='bar')
plt.title('Top 10 Teams by Average Player Rating')
plt.xlabel('Team')
plt.ylabel('Average Overall Rating')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
player_stats = pd.read_csv('player_stats.csv')
team_data = pd.read_csv('team_data.csv')

# First, merge the datasets on team_api_id
merged_data = pd.merge(
    player_stats,
    team_data,
    left_on='id',
    right_on='team_api_id',
    how='inner'
)

# Define leagues and their teams using team_short_name
leagues = {
    'Premier League': ['MUN', 'CHE', 'ARS', 'LIV', 'MCI', 'TOT'],
    'La Liga': ['REA', 'BAR', 'ATM', 'VAL', 'SEV'],
    'Bundesliga': ['BMU', 'DOR', 'LEV', 'S04', 'WOL'],
    'Serie A': ['JUV', 'INT', 'ACM', 'NAP', 'ROM']
}

# Create separate plots for each league
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.ravel()

for idx, (league, teams) in enumerate(leagues.items()):
    # Filter teams for each league
    league_data = merged_data[merged_data['team_short_name'].isin(teams)]
    
    # Calculate average ratings for teams in the league
    league_ratings = league_data.groupby('team_long_name')['overall_rating'].mean().sort_values(ascending=False)
    
    # Plot
    league_ratings.plot(kind='bar', ax=axes[idx], color='skyblue')
    axes[idx].set_title(f'Average Player Ratings in {league}')
    axes[idx].set_xlabel('Team')
    axes[idx].set_ylabel('Average Rating')
    axes[idx].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
attributes = ['overall_rating', 'potential', 'crossing', 'finishing', 'heading_accuracy', 'short_passing', 'volleys', 'dribbling', 'curve', 'free_kick_accuracy', 'long_passing', 'ball_control']

plt.figure(figsize=(12, 10))
sns.heatmap(merged_data[attributes].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Player Attributes')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['overall_rating'], merged_data['potential'], alpha=0.5)
plt.title('Overall Rating vs Potential')
plt.xlabel('Overall Rating')
plt.ylabel('Potential')
plt.show()
```
